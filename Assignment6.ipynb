```python
 BOKEH ASSIGNMENT
 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
import numpy as np

# Generate x-values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute y-values as the sine of x
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', plot_height=400, plot_width=600)

# Add a line renderer to the plot
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot
show(p)
```

```python
 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns
from bokeh.plotting import figure, show
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Generate random x and y values
n = 100
x = np.random.rand(n) * 10  # 100 random values between 0 and 10
y = np.random.rand(n) * 10  # 100 random values between 0 and 10

# Generate random sizes and colors for the markers
sizes = np.random.rand(n) * 20 + 10  # Sizes between 10 and 30
colors = np.random.choice(['red', 'green', 'blue', 'yellow', 'orange'], size=n)

# Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', plot_height=400, plot_width=600)

# Add a scatter plot with variable marker sizes and colors
p.scatter(x, y, size=sizes, color=colors, legend_field="color", fill_alpha=0.6)

# Show the plot
show(p)
```
